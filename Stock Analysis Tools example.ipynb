# Portfolio Analysis

```python
# using Corranalyzer to see the daily return correlation of each stock
from metaprodictor import Corranalyzer

ticker_group = ['DVN','CLR','MRO','FANG','CVE','TRGP','SSL','EOG','COP','IMO']
highlight = 0.78

b = Corranalyzer(ticker_group)
```

```python
# Display the Correlation Maxtrix 
b.matrixCorrl(highlight)
```

```python
# export csv file 
import pandas as pd
d = pd.DataFrame(b.matrix_csv())
d.to_csv('matrix.csv')
```

## Concise Summary of the Past Performance 
* Risk is the standard deviation of the ticker's close to close return.
* Upper/Lower Bounds are 2 SD away from the mean
* Windays = total positive return
* Profit = sum of positive daily % return)
* PnL = total profits - total losses

```python
# Ranking the risk(Daily return SD vs.expected return)
b.riskRank()
```
